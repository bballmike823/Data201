# Global Carbon Dioxide Emmissions

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/datasets/WDI/WDIreshaped.csv")
```

```python
df.columns.values
```

```python
world = df.loc[df.country_name == "World"]
df = df.loc[df.country_name != "World"]
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
late_years = df.loc[df.year > 2010]
early_years = df.loc[df.year < 2011]
```

```python
late_years.describe()
```

```python
early_years.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import seaborn as sns
```

```python
y = early_years.pop("co2_emissions").values
y.shape
```

```python
X = early_years.loc[:,"agriculture_land_area":"deaths"]
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 5)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
LinReg = LinearRegression()
LinReg.fit(X_train, y_train)
y_predicted = LinReg.predict(X_test)
print(mean_squared_error(y_test, y_predicted))
```

```python
print(r2_score(y_test,y_predicted))
```

```python
print(mean_squared_error(y_train, LinReg.predict(X_train)))
print(r2_score(y_train, LinReg.predict(X_train)))
```

```python
pd.DataFrame(
    {"attribute" : X.columns.values,
     "coefficient" : LinReg.coef_}
)
```

```python
plt.figure(figsize=(15,6))
plt.hist(y_predicted - y_test)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(y_test, y_predicted - y_test)
plt.show()
```
